```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('C:\\Users\\Madhesh\\ex1data1.txt', header = None) #read from dataset
y = data.iloc[:,0] # read first column
X = data.iloc[:,1] # read second column
m = len(y) # number of training example
```

```python
plt.scatter(X, y)
plt.xlabel('No.of positive cases from May 1 - Present (in 1000s)')
plt.ylabel('No. of Deaths (in 100s)')
plt.title('Tamilnadu Covid19 Death rate predictions')
plt.show()

X = X[:,np.newaxis]
y = y[:,np.newaxis]
theta = np.zeros([2,1])
print(X.shape)
```

```python
#finding hypothesis and cost function
def costfunction(X,y,theta):
    h = np.dot(X,theta) # 38 x 1
    J = (1/(2 * m)) * sum(np.power((h-y),2)) #scalar
    return J
#Gradient descent
def gradientdescent(X,y,theta,alpha,iterations):
    for i in range(iterations):
        temp = np.dot(X, theta) - y
        temp = np.dot(X.T,temp)
        theta = theta - (alpha/m) * temp
        #print(theta)
    return theta

alpha = 0.005
iterations = 50000
bias = np.ones((len(X),1))
X = np.append(bias,X, axis = 1)
theta = np.zeros([2,1])
J = costfunction(X,y,theta)
print("Before Gradient descent")
print(J)
theta = gradientdescent(X,y,theta,alpha,iterations)
print(theta)
```
